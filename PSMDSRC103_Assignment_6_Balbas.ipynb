```python
import random
from collections import Counter
import statistics as stats
import pandas as pd
random.seed(0)

glucose = [round(random.uniform(70, 200), 1) for _ in range(100)]
```

```python
#Compute Manually (No statistics, no pandas, no numpy)
#Mean
def mean(data):
    return sum(data) / len(data)

#Meadian
def median(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    mid = n // 2
    if n % 2 == 0:
        return (sorted_data[mid - 1] + sorted_data[mid]) / 2
    else:
        return sorted_data[mid]

#Mode
def mode(data):
    freq = Counter(data)
    max_freq = max(freq.values())
    modes = [k for k, v in freq.items() if v == max_freq]
    return modes[0] if len(modes) == 1 else modes

#Sample Variance
def sample_variance(data):
    m = mean(data)
    return sum((x - m) ** 2 for x in data) / (len(data) - 1)
def std_dev(data):
    return variance(data) ** 0.5

#Sample Standard Deviation
def sample_std_dev(data):
    return sample_variance(data) ** 0.5
```

```python
#Manual Calculations
manual_results = {
    "Mean": mean(glucose),
    "Median": median(glucose),
    "Mode": str(mode(glucose)),
    "Sample Variance": sample_variance(glucose),
    "Sample Std. Deviation": sample_std_dev(glucose)
}
```

```python
#Statistics Module
stats_results = {
    "Mean": stats.mean(glucose),
    "Median": stats.median(glucose),
    "Mode": str(stats.mode(glucose)),
    "Sample Variance": stats.variance(glucose),
    "Sample Std. Deviation": stats.stdev(glucose)
}
```

#Exercise 2

Using the previously generated data,
calculate the following statistics using the functions in the statistics module where appropriate.

```python
#RANGE
data_range = max(glucose) - min(glucose)
```

```python
#COEFFICIENT OF VARIATION
mean_value = stats.mean(glucose)
stdev_value = stats.stdev(glucose)
cv = (stdev_value / mean_value) * 100
```

```python
#INTERQUARTILE RANGE
quartiles = stats.quantiles(glucose, n=4)
q1 = quartiles[0]
q3 = quartiles[2]
iqr = q3 - q1
```

```python
#QUARTILE COEFFICIENT OF DISPERSION
qcd = (q3 - q1) / (q3 + q1)
```

```python
dispersion_results = {
    "Range": data_range,
    "Coefficient of Variation (%)": cv,
    "Interquartile Range (IQR)": iqr,
    "Quartile Coefficient of Dispersion (QCD)": qcd
}

df_dispersion = pd.DataFrame.from_dict(dispersion_results, orient="index", columns=["Value"])
df_dispersion.style.format("{:.4f}")
```
