Exercise 1: Read in the meteorite data from the Meteorite_Landings.csv file, rename
the mass (g) column to mass, and drop all the latitude and longitude
columns. Sort the result by mass in descending order.

```python
import pandas as pd

meteorites= pd.read_csv('/content/Meteorite_Landings.csv', nrows=5)
meteorites.head()
```

```python
meteorites =  meteorites.rename(columns={'mass (g)': 'mass'})
meteorites.head()
```

```python
mask = meteorites.columns.str.contains('reclat|reclong', regex=True)
columns_to_drop = meteorites.columns[mask]
columns_to_drop
```

```python
meteorites = meteorites.drop(columns=columns_to_drop)
meteorites.head()
```

```python
meteorites = meteorites.sort_values(by='mass', ascending=False)
meteorites.head()
```

Explanation:The dataset was cleaned by renaming the mass (g) column to mass for consistency and removing unnecessary latitude and longitude fields. Sorting by mass highlights the largest meteorites, which aids in quick comparative analysis.

Exercise 2: Using the meteorite data from the Meteorite_Landings.csv file, update the year column to only contain
the year, convert it to a numeric data type, and create a new column indicating whether the meteorite
was observed falling before 1970. Set the index to the id column and extract all the rows with IDs
between 10,036 and 10,040 (inclusive) with loc[].

 Hint 1: Use year.str.slice() to grab a substring.

 Hint 2: Make sure to sort the index before using loc[] to select the range.

Bonus: There's a data entry error in the year column. Can you find it? (Don't spend too much time on this.)

```python
meteorites['year'] = meteorites['year'].astype(str).str.slice(0, 4)
meteorites['year'] = pd.to_numeric(meteorites['year'], errors='coerce')
meteorites['before_1970'] = meteorites['year'] < 1970
meteorites = meteorites.set_index('id').sort_index()
meteorites.loc[100036:10040]
meteorites.head()
```

**Explanation:**This code standardizes the year column into numeric form, adds a Boolean field to identify pre-1970 meteorites, and uses the id as the index to enable direct record selection. This ensures chronological and structured data exploration.

1. Using the meteorite data from the Meteorite_Landings.csv file, create a pivot table that shows
both the number of meteorites and the 95th percentile of meteorite mass for those that were
found versus observed falling per year from 2005 through 2009 (inclusive). Hint: Be sure to
convert the year column to a number as we did in the previous exercise.
2. Using the meteorite data from the Meteorite_Landings.csv file, compare summary statistics of the
mass column for the meteorites that were found versus observed falling.

```python
meteorites['year'] = pd.to_numeric(meteorites['year'], errors='coerce')
recent = meteorites[(meteorites['year'] >= 2005) & (meteorites['year'] <= 2009)]
def pct95(x):
  return x.quantile(0.95)
pivot = pd.pivot_table(
    recent,
    index='year',
    columns='fall',
    values='mass',
    aggfunc=['count', pct95]
)
meteorites.head()
```

**Explanation:**A pivot table was created to show both the number of meteorites and the 95th percentile of mass between 20052009. This helps reveal trends in frequency and size distribution of meteorite falls over time.

Exercise 4: Using the taxi trip data in the 2019_Yellow_Taxi_Trip_Data.csv file, resample the data to an hourly
frequency based on the dropoff time. Calculate the total trip_distance, fare_amount, tolls_amount, and
tip_amount, then find the 5 hours with the most tips.

```python
import pandas as pd

taxi = pd.read_csv('/content/2019_Yellow_Taxi_Trip_Data.csv', nrows=5)
taxi.head()
```

```python
taxi['tpep_pickup_datetime'] = pd.to_datetime(taxi['tpep_pickup_datetime'])
taxi['tpep_dropoff_datetime'] = pd.to_datetime(taxi['tpep_dropoff_datetime'])
```

```python
taxi = taxi.set_index('tpep_dropoff_datetime')
hourly = taxi.resample('H')[['trip_distance', 'fare_amount', 'tolls_amount', 'tip_amount']].sum()
taxi.head()
```

```python
top_tips = hourly.sort_values('tip_amount', ascending=False).head(5)
```

```python
top_tips
```

**Explanation:**By resampling taxi trips hourly and aggregating totals, the code identifies the five hours with the highest tips. This approach highlights temporal patterns in earnings and rider generosity.
